- HDBSCAN: O(z log z + z d^2) - The first term represents the time complexity of constructing the minimum spanning tree, while the second term represents the time complexity of performing the density-based clustering.
- DBSCAN: O(z log z + z^2 d) - The first term represents the time complexity of constructing the k-D tree or ball tree, while the second term represents the time complexity of performing the density-based clustering.
- K-Means: O(I k z d) - The time complexity depends on the number of iterations (I), the number of clusters (k), and the size of the dataset (z) and number of dimensions (d).
- Mean Shift: O(z^2 d) - The time complexity depends on the size of the dataset (z) and number of dimensions (d).
